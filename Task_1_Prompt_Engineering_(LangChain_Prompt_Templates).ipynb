```python
from langchain_core.prompts import PromptTemplate

# Task 1: Replace Hardcoded Prompts
# Requirement: No f-strings, use PromptTemplate
template_v1 = "Tell me a joke about {topic}"
prompt_template = PromptTemplate(input_variables=["topic"], template=template_v1)

# Usage
print("Task 1 Output:")
print(prompt_template.format(topic="AI"))
```

```python
from langchain_core.prompts import PromptTemplate

# Task 2: Multi-Input Prompt System
multi_input_template = "Explain {topic} to {audience} in a {tone} tone."
prompt_multi = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=multi_input_template
)

# Test Cases
test_cases = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

print("Task 2 Outputs:")
for test in test_cases:
    print(prompt_multi.format(**test))
```

```python
from langchain_core.prompts import PromptTemplate

# Task 3: Prompt Variations Engine
teaching_prompt = PromptTemplate(input_variables=["topic"], template="Explain {topic} clearly step by step")
interview_prompt = PromptTemplate(input_variables=["topic"], template="Ask 3 questions about {topic}")
storytelling_prompt = PromptTemplate(input_variables=["topic"], template="Explain {topic} as a story")

topic = "Quantum Computing"

print("Task 3 Outputs:")
print("Teaching:", teaching_prompt.format(topic=topic))
print("Interview:", interview_prompt.format(topic=topic))
print("Storytelling:", storytelling_prompt.format(topic=topic))
```

```python
from langchain_core.prompts import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate

# Task 4: ChatPromptTemplate System
def get_chat_prompt(role, user_input):
    system_templates = {
        "teacher": "You are a helpful teacher.",
        "interviewer": "You are a professional interviewer.",
        "motivator": "You are an inspiring motivator."
    }

    system_message_prompt = SystemMessagePromptTemplate.from_template(system_templates.get(role, "You are a helpful assistant."))
    human_message_prompt = HumanMessagePromptTemplate.from_template("{text}")

    chat_prompt = ChatPromptTemplate.from_messages([system_message_prompt, human_message_prompt])
    return chat_prompt.format_prompt(text=user_input).to_messages()

print("Task 4 Output (Teacher Role):")
print(get_chat_prompt("teacher", "What is gravity?"))
```

```python
from langchain_core.prompts import PromptTemplate

# Task 5: Input Validation Layer
def validate_and_format(topic, audience, tone):
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    if audience not in valid_audiences:
        audience = "beginner"
    if tone not in valid_tones:
        tone = "formal"

    template = "Explain {topic} for {audience} in a {tone} style."
    prompt = PromptTemplate(input_variables=["topic", "audience", "tone"], template=template)
    return prompt.format(topic=topic, audience=audience, tone=tone)

# Task 6: Prompt Generator App
def prompt_generator_app(style, topic, audience, tone):
    styles = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style.",
        "interview": "Generate interview questions about {topic} for {audience} in a {tone} style.",
        "storytelling": "Tell a story about {topic} for {audience} in a {tone} style."
    }

    selected_template = styles.get(style, "Explain {topic} for {audience} in a {tone} style.")
    prompt = PromptTemplate(input_variables=["topic", "audience", "tone"], template=selected_template)
    return prompt.format(topic=topic, audience=audience, tone=tone)

print("Task 5 & 6 Outputs:")
print("Validated Prompt:", validate_and_format("Neural Networks", "beginner", "fun"))
print("App Prompt:", prompt_generator_app("storytelling", "Neural Networks", "beginners", "fun"))
```

```python
from langchain_core.prompts import PromptTemplate

# Task 7: Template Reusability Test
# Goal: Use ONE template with 5 different inputs
reusable_template = "Create a {style} summary about {topic}."
prompt_reusable = PromptTemplate(input_variables=["style", "topic"], template=reusable_template)

inputs = [
    {"style": "bullet-point", "topic": "Photosynthesis"},
    {"style": "one-sentence", "topic": "The French Revolution"},
    {"style": "technical", "topic": "Blockchain"},
    {"style": "poetic", "topic": "The Ocean"},
    {"style": "humorous", "topic": "Quantum Entanglement"}
]

print("Task 7 Outputs (Reusability Test):")
for item in inputs:
    print(prompt_reusable.format(**item))
```
